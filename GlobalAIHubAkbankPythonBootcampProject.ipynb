```python
import csv
from datetime import datetime
```

```python
class Pizza:
  description = ""
  price = int()
  def get_description(self):
    return self.description
  def get_cost(self):
    return self.price
  def add_sauce(self, sauce):
    self.description += f', {sauce.get_description()}'
    self.price += sauce.get_cost()
  def __init__(self):
    pass

  def __init__ (self, description):
    self.description = "Domates_Sosu, Mozerella_Peyniri, "+ description
  '''def __init__(self, price, description):
    self.price= price
    self.description = description
  def __init__(self, price):
    self.price= price
    self.description = "bos"
  def __init__(self, description):
    self.description = description
    self.price = 100'''

class BalliHardalliPizza(Pizza):
  def __init__ (self): 
    self.price = 170
    super().__init__("Sogan, Mantar, BalliHardalliSos")
class Margarita(Pizza):
  def __init__ (self):
    self.price = 155
    super().__init__("Feslegen")
class TurkPizza(Pizza):
  def __init__ (self):  
    self.price = 180
    super().__init__("Sucuk, Pastirma, Biber, Mantar")
class KantinPizza(Pizza):
  def __init__ (self):
    self.price = 155
    super().__init__("Sosis, Misir")
```

```python
class Decorator(Pizza):
  def __init__(self):
    pass


class Zeytin(Decorator):
    def __init__ (self):
      self.price = 10
      self.description = "Zeytin"
class Mantar(Decorator):
    def __init__ (self):
      self.price = 10
      self.description = "Mantar"
class KeciPeyniri(Decorator):
    def __init__ (self):
      self.price = 20
      self.description = "KeciPeyniri"
class Et(Decorator):
    def __init__ (self):
      self.price = 30
      self.description = "Et"
class Sogan(Decorator):
    def __init__ (self):
     self.price = 5
     self.description = "Sogan"
class Misir(Decorator):
    def __init__ (self):
      self.price = 5
      self.description = "Misir"
```

```python
def check_name():
  flag = True
  while flag:
    name = input('Lutfen isim, soyisim giriniz: ')
    for i in name:
      if i.isalpha() or i.isspace() :
        flag = False
        return name
      else:
        print('Sadece karakter girilebilir.')
```

```python
def check_tckn():
  flag = True
  while flag:
    tckn = input('Lutfen TC kimlik numaranizi giriniz: ')
    if tckn.isdigit() and len(tckn) == 11:
      flag = False
      return tckn
    else :
      print('TCKN 11 haneli ve sadece rakamlardan olusmalidir.')
      
```

```python
def check_credit_card_no():
  flag = True
  while flag:
    credit_card_number = input('Lutfen kredi karti numaranizi giriniz: ')
    if credit_card_number.isdigit() and len(credit_card_number) == 16:
      flag = False
      return credit_card_number
    else:
      print('Gecersiz kredi karti numarasi.')
```

```python
def check_password():
  flag = True
  while flag:
    password = input('Lutfen karti sifrenizi giriniz: ')
    if len(password) == 4 or len(password) == 6:
      flag = False
      return password
    else :
      print('Gecersiz sifre')
```

```python
def getUserInfos():
  name = check_name()
  tckn = check_tckn()
  ccn = check_credit_card_no()
  password = check_password()
  return name, tckn, ccn, password
```

```python
def create_csv():
  col_names=['isim', 'tckn', 'kredi karti numarasi', 'kredi karti sifresi', 'siparis aciklamasi', 'ücret', 'siparis saati']
  with open('Orders_Database.csv', 'w') as csv_obj:
    writer_object = csv.writer(csv_obj)
    writer_object.writerow(col_names)
    csv_obj.close()

create_csv()
```

```python
def get_time():
  now = datetime.now() 
  dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
  return dt_string
```

```python
def check_sauces(sauces):
  for i in sauces:
    if i.isdigit() == False:
      return False
    elif int(i)<11 and int(i)>16:
      return False
  return True
```

```python
pizza_array = [Margarita(), KantinPizza(), TurkPizza(), BalliHardalliPizza()]
sauce_dict = {'11' : Zeytin(), '12' : Mantar(), '13' : KeciPeyniri(), '14' : Et(), '15' : Sogan(), '16' : Misir()}
haveInfo = False

with open("/content/Menu.txt") as file: 
  menu = file.read()
  file.close()
  
exit = True
orders = []

while(exit):
  print(menu)
  pizza_code = input("Lutfen bir pizza numarasi giriniz: ")

  if(int(pizza_code)>=1 and int(pizza_code)<=4):
    pizza = pizza_array[int(pizza_code)-1]
    sauce_number_str = input("Lutfen istediginiz sos numaralarini arada bosluk olacak sekilde giriniz: ")
    sauce_number_str = sauce_number_str.split()
    sauce = check_sauces(sauce_number_str)
    if sauce == False:
      print('Hatali giris yapildi...')
      break

    for i in sauce_number_str:
      sauce = sauce_dict[i]
      pizza.add_sauce(sauce)
    onay = input(f'{pizza.get_description()} malzemelerini iceren pizzanin fiyati: {pizza.get_cost()}\n onayliyorsaniz Onay, Eğer onaylamıyorsaniz Ret yaziniz:')

    if onay == 'Ret':
      print('Siparis Reddedildi...')

    elif onay == 'Onay' and haveInfo == False:
      haveInfo = True
      name, tckn, ccn, password = getUserInfos()
      orders.append([name, tckn, ccn, password, pizza.get_description(), pizza.get_cost(), get_time()])

    elif onay == 'Onay' and haveInfo == True:
      orders.append([name, tckn, ccn, password, pizza.get_description(), pizza.get_cost(), get_time()])

    cikis = input('Siparisi tamamlamak icin "bitir" yaziniz ya da devam etmek icin enter tusuna basiniz: ')

    if cikis == 'bitir' and len(orders) != 0:
      print('Siparisiniz hazirlaniyor...')
      with open('Orders_Database.csv', 'a') as csv_obj:
        writer_object = csv.writer(csv_obj)
        for order in orders:
          writer_object.writerow(order)
        csv_obj.close()
      exit = False
    elif cikis == 'bitir':
        exit = False
    
  else:
    print("Hatali giris...")
    exit = False

```
